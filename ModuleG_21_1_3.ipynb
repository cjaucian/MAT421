# 21.1 Numerical Integration Problem Statement

To find the integral of a function, f(x), over an interval from a to b we can approximate the value. The function can be seen in the figure below.

The interval [a, b] is then discretized into a grid which contains n+1 points. The spacing of the points is determined by h = (b-a)/n. The individual points are set as xi, with x0 = a and xn = b. A subinterval is determined as [xi, xi+1]. The next two sections will approximate the area under the function by setting the subinterval to a shape for each method.

# 21.2 Riemanns Integral

One of the easiest methods for approximating the area under a function, is to use a rectangle area approximation. Each subinterval will create a rectangle, then the sum of the rectangles will create the approximate area of the integral. The width of the rectangle is given as: xi+1 - xi = h. The height of the rectangle is given as the function at x within the specified subinterval. The height of the rectangle can be given at the left endpoint, xi, or the right endpoint, xi+1.

This method is known as the Riemann Integral approximation.
integral2.jpg

Another method of the Rieman Integral approximation is the Midpoint Rule, which takes the height of the rectangle to be the midpoint of the subinterval. We determine the height by y = (xi+1 + xi)/2.

**Example**  Use the left Riemann Integral, right Riemann Integral, and Midpoint Rule to approximate the definite integral of sin(x)dx over the interval [0,pi] with 11 evenly spaced grid points. Compare this value to the exact value of 2.

```python
import numpy as np

a = 0
b = np.pi
n = 11
h = (b - a)/(n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 2 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

# 21.3 Trapezoid Rule

The Trapezoid Rule uses trapezoids within the subintervals to approximate the sum of the integrals. The corners of the trapezoid are given at (xi, 0), (xi+1, 0), (xi, f(xi)), and (xi+1, f(xi+1)). To find the area of the trapezoid we use: h*((f(xi) + f(xi+1))/2).

Many of the points in the Trapezoid Rule end up being counted twice due to the computation of the area of the trapezoid within each subinterval.

**Example**  Use the Trapezoid Rule to approximate the definite integral of sin(x)dx over the interval [0, pi] with 11 evenly spaced grid points. Compare this value ot the exact value of 2.

```python
from typing import IO
import numpy as np

a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])
err_trap = 2 - I_trap

print(I_trap)
print(err_trap)
```

Although double counting seems to occur, the Trapezoid Rule is as accurate as the Left or Right Riemann method.
